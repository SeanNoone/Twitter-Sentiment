```python
counter = 0


twitter_df = []

while counter < len(target_users):
    
    user = target_users[counter]

    for x in range(5):
  
        public_tweets = api.user_timeline(user, counter = 100, result_type = "recent")
    
        #print(public_tweets)
  
        for tweet in public_tweets:

            text = tweet["text"]
            compound = analyzer.polarity_scores(text)["compound"]
            pos = analyzer.polarity_scores(text)["pos"]
            neu = analyzer.polarity_scores(text)["neu"]
            neg = analyzer.polarity_scores(text)["neg"]
            tweets_ago = counter
        
        
            twitter_df.append({"News Outlet": (tweet['user']['name']),
                           "Compound": compound,
                           "Positive": pos,
                           "Negative": neu,
                           "Neutral": neg,})
        
    counter = counter + 1
    
```
